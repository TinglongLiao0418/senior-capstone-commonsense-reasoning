```python
# Filter out non-english entries and make start/end readable
from tqdm import tqdm
rows = 34074917

src_path = "conceptnet-assertions-5.7.0.csv"
src_file = open(src_path, 'r')
tar_path = "conceptnet.csv"
tar_file = open(tar_path, 'w')

with tqdm(total=rows) as pbar:
    while True:
        pbar.update(1)
        line = src_file.readline()
        
        if not line: 
            break
        
        
        _, r, s, e, _ = line.split('\t')
        
        if s.split('/')[2] != 'en' or e.split('/')[2] != 'en':
            continue
        
        r = r.split('/')[2]
        s = s.split('/')[3]
        e = e.split('/')[3]
        new_line = r + '\t' + s + '\t' + e + '\n'
        
        tar_file.write(new_line)
        


src_file.close()
tar_file.close()
```
